```python
# ライブラリーの読み込み
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# tipsデータセットの読み込み
tips = sns.load_dataset("tips")

# tipsデータセットの中身を表示
tips
```

```python
# 特徴量の分布や相関をグラフ表示
sns.pairplot(data=tips)
```

```python
# total_billを説明変数X, tipを目的変数yとする
X = tips[["total_bill"]].values
y = tips["tip"].values
```

```python
# 線形回帰クラスをインスタンス化
lr = LinearRegression()

# データセットを用いて学習
lr.fit(X, y)

# フィッティングで得られた直線
y_pred = lr.predict(X)
```

```python
# データ点と得られた直線をプロット
plt.scatter(X, y)
plt.plot(X, y_pred, color="black")
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.show()
```

```python
# 直線の傾き(Slope)と切片(Intercept)を表示
print(f'傾き: {lr.coef_[0]:.3f}')
print(f'切片: {lr.intercept_:.3f}')

# 平均二乗誤差を計算し、表示
mse_train = mean_squared_error(y, y_pred)
print(f'誤差: {mse_train:.3f}')
```
